1 각 피드백 중에서 가장 긴 피드백을 작성한 UserID를 찾고 해당 UserID가 주문을 몇 번 했는지 구하시오 (3)

2 주어진 데이터에서 '제품'이라는 단어가 가장 많이 포함된 카테고리(그룹)을 찾고, 해당 카테고리에 속한 피드백들의 평균 배송 시간(분)을 구하시오. (9048.75)

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/e-commerce.csv")
df['len'] = df['Feedback'].str.len()
df2 = df.sort_values('len', ascending = False).reset_index(drop = True)
df2.head()
maxlen_userid = df2.iloc[0, 4]
print(maxlen_userid) #가장긴피드백

cond = df['UserID'] == maxlen_userid
order_cnt = len(df[cond])
print(order_cnt) #해당userid의 주문수

cond2 = df['Feedback'].str.contains("제품")
df3 = df[cond2]
df3_cate_max = df3['Category'].value_counts().idxmax()
print(df3_cate_max) #제품 이야기 가장 많이 나온 카테고리는 서비스

cond3 = df['Category'] == df3_cate_max
df4 = df[cond3]
df4['OrderDate'] = pd.to_datetime(df4['OrderDate'])
df4['ArrivalDate'] = pd.to_datetime(df4['ArrivalDate'])
df4['timegap'] = (df4['ArrivalDate'] - df4['OrderDate']).dt.total_seconds()/60
avg = df4['timegap'].mean()
print(avg)
```

```python
# 피드백의 길이를 계산하고 새로운 컬럼 'len'에 저장
df['len'] = df['Feedback'].str.len()

# 가장 긴 피드백을 가진 행의 인덱스를 찾고 해당 UserID를 추출
idx = df['len'].idxmax()
user = df.iloc[idx]['UserID']

# 해당 UserID로 필터링하여 주문 횟수 계산
cond = df['UserID'] == user
len(df[cond])

# 3
```

```python
# 각 카테고리별로 '제품'이라는 단어가 포함된 피드백의 수를 계산
category_counts = df.groupby('Category')['Feedback'].apply(lambda x: x.str.contains('제품').sum())

# '제품'이라는 단어가 가장 많이 포함된 카테고리 찾기
max_category = category_counts.idxmax()

# 배송 시간 계산을 위해 'ArrivalDate'와 'OrderDate' 열을 datetime 형식으로 변환
df['ArrivalDate'] = pd.to_datetime(df['ArrivalDate'])
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# 배송 시간(분) 계산
df['배송시간(분)'] = (df['ArrivalDate'] - df['OrderDate']).dt.total_seconds() / 60

# 찾은 카테고리의 평균 배송 시간(분) 계산
mean_delivery_time = df[df['Category'] == max_category]['배송시간(분)'].mean()

print(mean_delivery_time)
```
