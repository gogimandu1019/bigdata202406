## 데이터 전처리 & 피처엔지니어링

```python
train.head()
```

```python
#name, hostname, host_id, last_review 삭제
cols = ['name','host_id','last_review', 'host_name']
print(train.shape)
train = train.drop(cols, axis = 1)
test = test.drop(cols, axis = 1)
print(train.shape)
```

```python
# reviews_per_month 결측치 0으로 채우기
train['reviews_per_month'] = train['reviews_per_month'].fillna(0)
test['reviews_per_month'] = test['reviews_per_month'].fillna(0)
train.isnull().sum()
```

```python
train = train.drop('id', axis = 1)
test_id = test.pop('id')
test.head()
```

```python
train.info()
```

```python
#object 3개 숫자로 변경-> label 인코딩 : 원핫인코딩 하기에는 개수가 너무 많음
from sklearn.preprocessing import LabelEncoder

cols = ['neighbourhood_group','neighbourhood', 'room_type']

for col in cols:
  le = LabelEncoder()
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])

train[cols]
```

## 검증 데이터 분리

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(train.drop('price', axis = 1), train['price'], test_size = 0.15, random_state = 2022 )
X_val.head()
```

## 모델 & 평가

```python
# 평가
import numpy as np
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
def rmse(y_test, y_pred): #RMSE
    return np.sqrt(mean_squared_error(y_test, y_pred))

def rmsle(y_test, y_pred): #RMSLE
    return np.sqrt(np.mean(np.power(np.log1p(y_test) - np.log1p(y_pred), 2)))

def mape(y_test, y_pred): #MAPE
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
# 선형회귀
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_tr, y_tr)
pred = model.predict(X_val)

print("r2: " , r2_score(y_val, pred))
print("mae: ", mean_absolute_error(y_val, pred))
print("mse: ", mean_squared_error(y_val, pred))
print("rmse: ", rmse(y_val, pred))
print("rmsle: ", rmsle(y_val, pred))
print("mape: ", mape(y_val, pred))
```

```python
#릿지
from sklearn.linear_model import Ridge

model = Ridge()
model.fit(X_tr, y_tr)
pred = model.predict(X_val)

print("r2: " , r2_score(y_val, pred))
print("mae: ", mean_absolute_error(y_val, pred))
print("mse: ", mean_squared_error(y_val, pred))
print("rmse: ", rmse(y_val, pred))
print("rmsle: ", rmsle(y_val, pred))
print("mape: ", mape(y_val, pred))
```
