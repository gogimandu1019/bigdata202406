```python
import pandas as pd

# 데이터 불러오기 (datatime컬럼 지정)
df = pd.read_csv("../input/bigdatacertificationkr/basic2.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.info()

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['weekday'] = df['Date'].dt.weekday #요일 : 월요일이 0 ~ 일요일이 6
df.info()
df.head()
cond1 = (df['year'] == 2022)
cond2 = (df['month'] == 5)
cond3 = (df['weekday'] >= 5) #주말
cond4 = (df['weekday'] < 5) #평일
df_weekend = df[cond1 & cond2 & cond3]
df_weekday = df[cond1 & cond2 & cond4]
sales_weekday =df_weekday['Sales'].mean()
print(sales_weekday)
sales_weekend = df_weekend['Sales'].mean()
print(sales_weekend)
answer = round(sales_weekend - sales_weekday, 2)
print(answer)
```
