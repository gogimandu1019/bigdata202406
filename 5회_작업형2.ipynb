### 답안 제출 참고
- pd.read_csv('result.csv') 로 제출 코드 확인


```python
import pandas as pd
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train.shape, test.shape #((3759, 9), (1617, 8))

train.head() #price예측 : 회귀
#test.head()
#train.info() #object cols = model, transmission, fuelType
#test.info()
train.describe() #연도에 2060이 조금 이상한데 ... 이거 바꿨어야 했나
test.describe()
test.describe(include = "O") #model unique 19? 겟더미 괜찮을까
test.isnull().sum() #train test 모두 결측치 없음

train = pd.get_dummies(train)
train.shape #(3759, 31)
test = pd.get_dummies(test)
test.shape #(1617, 30)

target = train.pop("price")
train.shape #(3759, 30)

#train-test분리
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)
x_tr.shape, x_val.shape, y_tr.shape, y_val.shape #((3007, 30), (752, 30), (3007,), (752,))

#randomforestregressor
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state = 0)
model.fit(x_tr, y_tr)
pred = model.predict(x_val)

from sklearn.metrics import mean_squared_error
def getRmse(a,b):
  ret = 0
  ret = mean_squared_error(a, b) **0.5
  return ret

getRmse(y_val, pred) #1341.943732584459

pred2 = model.predict(test)

submission = pd.DataFrame({"pred" : pred2}).to_csv("result.csv", index = False)
pd.read_csv("result.csv")
```
