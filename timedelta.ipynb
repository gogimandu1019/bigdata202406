1. 세션의 지속 시간을 분으로 계산하고 가장 긴 지속시간을 출력하시오(반올림 후 총 분만 출력)
2. 가장 많이 머무른 Page를 찾고 그 페이지의 머문 평균 시간을 구하시오 (반올림 후 총 시간만 출력) 
3. 사용자들이 가장 활발히 활동하는 시간대(예: 새벽, 오전, 오후, 저녁)를 분석하세요. 이를 위해 하루를 4개의 시간대로 나누고 각 시간대별로 가장 많이 시작된 세션의 수를 계산하고, 그 중에 가장 많은 세션 수를 출력하시오
    - 새벽: 0시 부터 6시 전
    - 오전: 6시 부터 12시 전
    - 오후: 12 부터 18시 전
    - 저녁: 18시 부터 0시 전

4. user가 가장 많이 접속 했던 날짜를 출력하시오. (예, 2023-02-17)

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/website.csv")
```

```python
#df.info()
df['diff'] = pd.to_datetime(df['EndTime']) - pd.to_datetime(df['StartTime'])
df['diff'] = round(df['diff'].dt.total_seconds()/60, 0)
maxx = df['diff'].max()
print(maxx) #300.0
```

```python
#가장 많이 머무른 Page를 찾고 그 페이지의 머문 평균 시간을 구하시오 (반올림 후 총 시간만 출력)
df2 = df.groupby('Page')['diff'].mean()
df2
maxpage = df2.max()
print(int(round(maxpage/60, 0))) #3
```

```python
# 사용자들이 가장 활발히 활동하는 시간대(예: 새벽, 오전, 오후, 저녁)를 분석하세요. 이를 위해 하루를 4개의 시간대로 나누고 
# 각 시간대별로 가장 많이 시작된 세션의 수를 계산하고, 그 중에 가장 많은 세션 수를 출력하시오

# 새벽: 0시 부터 6시 전
# 오전: 6시 부터 12시 전
# 오후: 12 부터 18시 전
# 저녁: 18시 부터 0시 전
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/website.csv")
df

df['startHour'] = pd.to_datetime(df['StartTime']).dt.hour
df.head()

def chkTime(hour):
    ret = ""
    if hour >= 0 and hour < 6:
        ret = "새벽"
    elif hour >= 6 and hour < 12:
        ret = "오전"
    elif hour >= 12 and hour < 18:
        ret = "오후"
    else:
        ret = "저녁"
    return ret

df['timename'] = df['startHour'].apply(chkTime)
df2 = df.groupby('timename')['UserID'].count()
print(df2.max())
```

```python
#user가 가장 많이 접속 했던 날짜를 출력하시오. (예, 2023-02-17)
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/website.csv")
df['StartTime'] = pd.to_datetime(df['StartTime'])
df['EndTime'] = pd.to_datetime(df['EndTime'])
df['startDate'] = df['StartTime'].dt.date
df.head()

df2 = df.groupby('startDate')['UserID'].count().sort_values(ascending = False)
print(df2.idxmax())
```

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/website.csv")
```

```python
# 1
# 날짜 형식으로 변환
df['StartTime'] = pd.to_datetime(df['StartTime'])
df['EndTime'] = pd.to_datetime(df['EndTime'])
df['SessionDuration'] = (df['EndTime'] - df['StartTime']).dt.total_seconds() / 60 
print(df['SessionDuration'].max())
```

```python
# 2
most_visited_page = df['Page'].value_counts().idxmax()
cond = df['Page'] == most_visited_page
result = df[cond]['SessionDuration'].mean()
print(round(result/60))
```

```python
# 3
def time_slot(hour):
    if 0 <= hour < 6:
        return "새벽"
    elif 6 <= hour < 12:
        return "오전"
    elif 12 <= hour < 18:
        return "오후"
    else:
        return "저녁"

df['TimeSlot'] = df['StartTime'].dt.hour.apply(time_slot)
print(df['TimeSlot'].value_counts()[0])
```

```python
# 4
result = df['StartTime'].dt.date.value_counts().index[0]
print(result.strftime('%Y-%m-%d'))
```
