```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv')
df.shape
df
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e'
}

df['newIncome']= df['Income_Category'].map(lambda x : dic[x])
ans = df['newIncome']
ans
```

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':
        return 'b'
    elif x =='$60K - $80K':
        return 'c'
    elif x =='$80K - $120K':
        return 'd'
    elif x =='$120K +' :
        return 'e'


df['newIncome'] = df['Income_Category'].apply(changeCategory)
ans = df['newIncome']
ans
```

```python
def f(x):
  ret = x // 10 * 10
  return ret

df['AgeState'] = df['Customer_Age'].apply(f)
ans = df['AgeState'].value_counts().sort_index()
ans
```

```python
def f(x):
  if "Graduate" in x:
    ret = 1
  else:
    ret = 0
  return ret


df['newEduLevel'] = df['Education_Level'].apply(f)
ans = df['newEduLevel'].value_counts()
ans
```

```python
def f(x):
  if x >= 4500:
    ret = 1
  else:
    ret = 0
  return ret

df['newLimit'] = df['Credit_Limit'].apply(f)
ans = df['newLimit'].value_counts()
ans
```

```python
def check(x):
  if x['Marital_Status'] == "Married" and x['Card_Category'] == "Platinum":
    ret = 1
  else:
    ret = 0
  return ret

df['newState'] = df.apply(check, axis = 1)
ans = df['newState'].value_counts()
ans
```

```python
def chk(x):
  if x == "M":
    ret = "male"
  elif x == "F":
    ret = "female"
  return ret

df['Gender'] = df['Gender'].apply(chk)
ans = df['Gender'].value_counts()
ans
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv")
df.info()
```

```python
df['Yr_Mo_Dy'].dt.year.unique()
```

```python
def fix_century(x):
  import datetime

  if x.year >= 2061:
    year = x.year-100
  else:
    year = x.year

  return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
df
```

```python
ans = df.groupby(df['Yr_Mo_Dy'].dt.year).mean()
ans
```

```python
df['weekday'] = df['Yr_Mo_Dy'].dt.weekday
df['weekday'].to_frame()
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
df.head()
```

```python
def f(x):
  import datetime
  date = x.split(":")[0]
  hour = x.split(":")[1]

  if hour == "24":
    hour = '00:00:00'
    ret = pd.to_datetime(date + " " + hour) + datetime.timedelta(days = 1)
  else:
    hour = hour + ":00:00"
    ret = pd.to_datetime(date + " " + hour)

  return ret

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(f)
df.head()
```

```python
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df.head()
```
