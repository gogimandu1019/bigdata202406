```python
import pandas as pd
df = pd.DataFrame({'A': ['블루베리 스무디', '딸기 스무디', '딸기 바나나 스무디'],
                   'B': [10, 20, 30],
                   'C': ['추천/신메뉴', '신메뉴/할인', '사이즈업/추천'],
                   'D': ['ab cd', 'AB CD', 'ab cd']
                   })
df
```

```python
df['A'] = df['A'].replace('스무디','에이드') #replace 는 전체가 일치해야 변경
df
```

```python
df['A'] = df['A'].str.replace('스무디','에이드') #str.replace는 일부 일치 시 변경, 숫자에는 사용불가
df
```

```python
#replace데이터프레임
df.replace('블루베리 에이드','청포도에이드')
```

```python
df['B'] = df['B'].replace(10, 100)
df
```
