```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
pd.set_option("display.max_columns" , None)

train = pd.read_csv("/kaggle/input/playground-series-s4e6/train.csv")
test = pd.read_csv("/kaggle/input/playground-series-s4e6/test.csv")


# train.info()
# test.info()

# train.describe()
# train.describe(include = "O")
# train.isnull().sum()

# test.describe()
# test.isnull().sum()

#print(train.shape, test.shape) #(76518, 38) (51012, 37)
train = train.drop("id", axis = 1)
target = train.pop("Target")
test_id = test.pop("id")
#print(train.shape, test.shape)

#minmax scaler
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()

train = mms.fit_transform(train)
test = mms.transform(test)

#train test split
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)
#print(x_tr.shape, x_val.shape, y_tr.shape, y_val.shape) #(61214, 36) (15304, 36) (61214,) (15304,)

#randomforest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 0)
model.fit(x_tr, y_tr)
pred = model.predict(x_val)
pred

from sklearn.metrics import accuracy_score
score = accuracy_score(y_val, pred)
#print(score) #스케일링 했을 때-0.8218112911657083, 스케일링 안 했을때 - 0.8214845791949817 : 스케일링 하는게 0.004 정도 낫다는 결론

pred = model.predict(test)

submit = pd.DataFrame({'id' : test_id, 'Target' : pred}).to_csv("submission.csv", index= False)

pd.read_csv("submission.csv")
```
