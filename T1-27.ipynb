## T1-27. 바닐라라떼 5점, 카페라떼 3점, 아메리카노 2점, 나머지 0점이다 총 메뉴의 점수를 더한 값은?

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/payment.csv")
df

df['menu'] = df['menu'].str.replace(" ", "")
df

cond1 = df['menu'] == "바닐라라떼"
cond2 = df['menu'] == "카페라떼"
cond3 = df['menu'] == "아메리카노"

score = len(df[cond1]) * 5 + len(df[cond2]) * 3 + len(df[cond3]) * 2
print(score)
```
