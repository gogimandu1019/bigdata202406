```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
pd.set_option('display.max_columns', None)

train = pd.read_csv("/kaggle/input/smoker-binary-class/train.csv")
#train.head()
test = pd.read_csv("/kaggle/input/smoker-binary-class/test.csv")
#test.head()

#train.info()
#test.info()

#train.isnull().sum()
#test.isnull().sum()

#train.describe()
#test.describe()

train.sort_values('waist(cm)', ascending = True)
#허리둘레 1은 누가봐도 이상값으로 보임 : 제거
cond = (train['waist(cm)'] == 1.0)
train = train[~cond]
#train.shape

target = train.pop("smoking")
#target
#target.head()
test_id = test.pop("id")
#test_id.head()
train = train.drop("id", axis = 1)

#train_test_split
from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)
#x_tr, x_val, y_tr, y_val
#x_tr.shape, x_val.shape, y_tr.shape, y_val.shape ((11999, 22), (3000, 22), (11999,), (3000,))

#randomforestClassifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 0)
model.fit(x_tr, y_tr)
pred = model.predict_proba(x_val)

from sklearn.metrics import roc_auc_score
ras = roc_auc_score(y_val, pred[:, 1])
ras #0.8854137470804138

pred = model.predict_proba(test)
submission = pd.DataFrame({"id" : test_id, "smoking" : pred[:,1]}).to_csv("submission.csv", index = False)
pd.read_csv("submission.csv")
```
