```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))


train = pd.read_csv('/kaggle/input/wine-qualit-dataset/train.csv')
test = pd.read_csv('/kaggle/input/wine-qualit-dataset/test.csv')\

#train.shape, test.shape #id 컬럼 드랍/ 분리 필요, target = quality, 15000개로 10000개 예측 : 테스트 사이즈 0.2 는 되어야 할거같고. 결측치 없고 범주형 없고

#일단 하던대로 전처리
target = train.pop('quality')
test_id = test.pop('id')
train = train.drop('id', axis = 1)

#train.shape, test.shape, target.shape, test_id.shape

from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(train, target, test_size = 0.2, random_state = 0)
#x_tr.shape, x_val.shape, y_tr.shape, y_val.shape #((12000, 11), (3000, 11), (12000,), (3000,))

#target데이터가 quality 분류일 것 같다: 점수들이 죄다 정수형임.
from sklearn.ensemble import RandomForestClassifier
#help(RandomForestClassifier)
model1 = RandomForestClassifier(random_state = 0, n_estimators = 150, max_depth = 10)

model1.fit(x_tr, y_tr)
pred = model1.predict(x_val)

#쿼드릭가중카파 : 이게 뭔지 모르겠어 구현을 못 하겠음, 일단 accuracy_score 사용
from sklearn.metrics import accuracy_score
score = accuracy_score(y_val, pred)
print(score) #0.5306666666666666 -> 튜닝하고 나니 0.5496666666666666


pred2 = model1.predict(test)

submission = pd.DataFrame({'id': test_id, 'quality' : pred2}).to_csv("submission.csv", index = False)

pd.read_csv("submission.csv")

```
